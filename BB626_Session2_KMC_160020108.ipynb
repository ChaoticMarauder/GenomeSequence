```
#Question-1
#Part A
from math import *
from random import *
import numpy as np
import matplotlib.pyplot as plt

Kp=10 #polymerisation rate
Kd=5 #depolymerisation rate

t=0 #initial time

num_iter=10000 #number of iterations

R=Kp+Kd

ts=np.empty(num_iter)
Ls=np.empty(num_iter)

L=1
i=0
for i in range(num_iter):
  ts[i]=t
  Ls[i]=L
  
  u1=uniform(0,1)
  if u1*R<=Kd:
    if L>0:
      L=L-1
  else:
    L=L+1
  
  u2=uniform(0,1)
  dt=-log(u2)/R
  t=t+dt
  
growth_rate=(L-Ls[0])/t
print("Growth rate", growth_rate, sep=':')

plt.figure(1)
plt.plot(ts, Ls, label='polymer length')
legend=plt.legend(loc='upper right', shadow=True)
plt.xlabel('time (s)')
plt.ylabel('polymer length')
plt.grid('True')
plt.show()   
```

```
#Question-3
#Part A
from math import *
from random import *
import numpy as np
import matplotlib.pyplot as plt

Rb_0=2 #binding rate
Rd_0=2 #unbinding rate

t=0 #initial time
N=1000

num_iter=20000 #number of iterations

R=Rb_0+Rd_0

B_0=np.zeros(N)

ts=np.empty(num_iter)
Bs=np.empty(num_iter)


for i in range(num_iter):
  ts[i]=t
  Bs[i]=np.sum(B_0)
  Rb=Rb_0*(N-Bs[i])
  Rd=Rd_0*Bs[i]
  R=Rb+Rd
  n=randint(0,N-1) 
  u1=uniform(0,1)
  if u1*R<=Rd:
    B_0[n]=0
  else:
    B_0[n]=1
  
  u2=uniform(0,1)
  dt=-log(u2)/R
  t=t+dt

   
#growth_rate=(L-Ls[0])/t
#print("Growth rate", growth_rate, sep=':')

plt.figure(1)
plt.plot(ts, Bs, label='Bound proteins')
legend=plt.legend(loc='lower right', shadow=True)
plt.xlabel('time (sec)')
plt.ylabel('number of bound protein')
plt.grid('True')
plt.show()   
```
